```c++
#include <iostream>
#include <cmath>
using namespace std;

{
    int radius;
    float pi = 3.14;
    float volume;

    cout << "Enter the radius of the sphere: ";
    cin >> radius;
    cout<<radius<<endl;

    //pow is a function that takes radius/base as 1st arg and exponent as the 2nd arg
    volume = (4.0/3.0) * pi * pow(radius, 3);

    cout << "Volume is: " << volume << endl;
}
```
