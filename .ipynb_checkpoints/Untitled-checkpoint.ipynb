```c++


// function to compute perimeter
int perimeter(int side) {
    return side * 4;
}

// function to compute area
int area(int side) {
    return side * side;
}

```

```c++
#include <iostream>
using namespace std;
{
    int side;

    cout << "Enter the side length of the square: ";
    cin >> side;

    cout << "Perimeter of the square: " << perimeter(side) << endl;
    cout << "Area of the square: " << area(side) << endl;
}
```
