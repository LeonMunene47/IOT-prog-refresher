```c++
#include <iostream>
using namespace std;

{
    int side;
    cout << "Enter side: ";
    cin >> side;

    int perimeter(int side) {
        int result = side * 4;
        return result;
    }

    int area(int side) {
        int result = side * side;
        return result;   
    }

    cout << perimeter(side) << endl;
    cout << area(side) << endl;
}
```
